```python
import numpy as np
import matplotlib.pyplot as plt #this import statement is used to import lib that plots graph
height=np.random.normal(140,20,1000)
plt.figure(dpi=200)
plt.title('Height Distribution')
plt.xlabel('height(cm)')
plt.ylabel('Frequency')
plt.hist(height,bins=30,ec='gold')
```

## Difference between interpreter and compiler
      Interpreter is used to run the code directly without compiling it. Whereas a compiler compiles the code first then run the code. After compilation it shows the error we have done in our code.
